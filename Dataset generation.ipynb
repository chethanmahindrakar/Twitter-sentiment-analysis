```python
#Dataset generation
#Take only 50,000 postive tweets and 50,000 negative tweets.
import numpy as np
import pandas as pd
import csv
import json
import random

#Path variables
dataset_path = "C:/Users/cheth/Anaconda Python/Twitter Sentiment Analysis/dataset/original_dataset.csv"

entire_dataset = "C:/Users/cheth/Anaconda Python/Twitter Sentiment Analysis/dataset/short_dataset.csv"
store_train_path = "C:/Users/cheth/Anaconda Python/Twitter Sentiment Analysis/dataset/training.csv"
store_test_path = "C:/Users/cheth/Anaconda Python/Twitter Sentiment Analysis/dataset/testing.csv"
```

```python
#Reading the dataset
cols = ['sentiment','id','date','query_string','user','text']
data = pd.read_csv(dataset_path,encoding='latin1',names=cols) 
```

```python
data
```

```python
#Generating the entire dataset - 120000 rows

#Reading 60000 random rows with negetive label
negetive_data = data[data.sentiment == 0]
rows = np.random.choice(negetive_data.index.values, 60000,replace= False)
df1 = negetive_data.loc[rows]

#Reading 60000 random rows with positive label
positive_data = data[data.sentiment == 4]
rows = np.random.choice(positive_data.index.values, 60000,replace= False)
df2 = positive_data.loc[rows]
```

```python
#Writting into the enitire_dataset csv file
df3 = pd.concat([df1, df2], axis=0)
df3.to_csv(entire_dataset, header=False)

del data , negetive_data , positive_data, df1,df2,df3
```

```python
#Generating the training and testing dataset - (80:20)
temp = pd.read_csv(entire_dataset,encoding='latin1',names=cols) 
temp
```

```python
import numpy as np
#Reading 50000 random rows with negetive label
negetive_data = temp[temp.sentiment == 0]
rows1 = np.random.choice(negetive_data.index.values, 10000,replace= False)
df1 = negetive_data.loc[rows1]

#Reading 50000 random rows with positive label
positive_data = temp[temp.sentiment == 4]
rows1 = np.random.choice(positive_data.index.values, 10000,replace= False)
df2 = positive_data.loc[rows1]

df1
df2
```

```python
#Concatinating the two datasets
df3 = pd.concat([df1, df2], axis=0)
#Writting into the testing_dataset csv file
df3.to_csv(store_test_path, header=False)

#Removing the rows from the dataset that have already been taken for the testing dataset
df4 = pd.concat([temp, df3, df3]).drop_duplicates(keep=False)
df4
#Writting into the training_dataset csv file
df4.to_csv(store_train_path, header=False)

#del df4,temp,negetive_data,positive_data,df3,df4
```
